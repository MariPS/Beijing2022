```python
import pandas as pd
athletes = pd.read_csv("datasets/athletes.csv")
curling_results = pd.read_csv("datasets/curling_results.csv")
entries_discipline = pd.read_csv("datasets/entries_discipline.csv")
hockey_results = pd.read_csv("datasets/hockey_results.csv")
medals_total = pd.read_csv("datasets/medals_total.csv")
medals = pd.read_csv("datasets/medals.csv")
```

```python
athletes.columns[athletes.isna().any()].tolist()
```

```python
pd.concat(g for _, g in athletes.groupby("name") if len(g) > 1)
```

Nel dataset sono presenti più tuple con lo stesso nome dell'atleta. Questo perché un atleta può gareggiare in più discipline. La chiave della tuple è data dalla combinazione <code>name</code>+<code>discipline</code>.

```python
pd.concat(g for _, g in athletes.groupby(["name","discipline"]) if len(g) > 1)
```
